```python
import pandas as pd

df = pd.read_csv("~/Downloads/train.csv")
df.head(10)
```

```python
import numpy as np
cols = ["manufactured", "curb_weight", "power", "engine_cap", "no_of_owners", "depreciation", "coe", "road_tax", "dereg_value", "mileage", "omv", "arf", "indicative_price", "price"]
df[cols] = df[cols].fillna(0).astype(int)
df["manufactured"] = df["manufactured"].astype(str)
df["manufactured"] = df["manufactured"].replace({"0": None})
df = df.replace({np.nan: None})
df.head(10)
```

```python
df["title"] = '"' + df["title"] + '"'
df["description"] = df["description"].str.strip('"')
df["description"] = '"' + df["description"] + '"'
df["category"] = '"' + df["category"] + '"'
df["features"] = df["features"].str.strip('"')
df["features"] = '"' + df["features"] + '"'
df["accessories"] = df["accessories"].str.strip('"')
df["accessories"] = '"' + df["accessories"] + '"'
df.head(10)
```

```python
df.to_csv("~/Downloads/train_preprocessed.csv", index=False)
```
